```python
from typing import TypedDict
from langgraph.graph import StateGraph, START, END
from langgraph.checkpoint.memory import MemorySaver  
```

```python
def ClassState(TypedDict):
    count : int
    message : str
```

```python
def increment(state : ClassState)->ClassState:
    new_count = state['count']+1
    log = state['message']+[f'Increment value{new_count}']

    return {'count':new_count,'message':log}
```

```python
builder  = StateGraph(ClassState)

```

```python
builder.add_node('increment',increment)
```

```python
builder.add_edge(START,'increment')
builder.add_edge('increment',END)
memory = MemorySaver()
graph = builder.compile(checkpointer=memory)
```

```python
print("Graph compiled with MemorySaver checkpointer.")
print("The graph will now remember state between calls (within the same thread_id).")
```
